```python
import numpy as np
import uproot
from helper import *
```

```python
# signal parameters
mu_sig = 125
sigma_sig = 17
yield_sig = 30

# background parameters
mu_bkg = 91.2
sigma_bkg = 15
yield_bkg = 500

# experiment limits
xmin, xmax = 60, 160
bins = np.linspace(xmin, xmax, 21)

# For the systematic variations:
rel_sigma_var = 0.05
rel_mu_var = 0.02
yield_scale = 1.2

n_mc = int(1e6)
```

```python
# do mc generation for all variations and plot
(
    h_sig_nom, h_bkg_nom,
    h_sig_up_1, h_bkg_up_1, h_sig_down_1, h_bkg_down_1,
    h_sig_up_2, h_bkg_up_2, h_sig_down_2, h_bkg_down_2,
    h_sig_up_3, h_bkg_up_3, h_sig_down_3, h_bkg_down_3,
    h_sig_up_4, h_bkg_up_4, h_sig_down_4, h_bkg_down_4,
) = generate_all_mc_variations(
    n_mc, mu_sig, sigma_sig, mu_bkg, sigma_bkg, yield_sig, yield_bkg,
    rel_sigma_var, rel_mu_var, yield_scale, bins=np.linspace(60, 160, 21),
)

h_sig_up = np.stack([h_sig_up_1, h_sig_up_2, h_sig_up_3, h_sig_up_4])
h_bkg_up = np.stack([h_bkg_up_1, h_bkg_up_2, h_bkg_up_3, h_bkg_up_4])
h_sig_down = np.stack([h_sig_down_1, h_sig_down_2, h_sig_down_3, h_sig_down_4])
h_bkg_down = np.stack([h_bkg_down_1, h_bkg_down_2, h_bkg_down_3, h_bkg_down_4])

for i in range(4):
    plot_systematic(h_sig_nom, h_bkg_nom, h_sig_up[i], h_bkg_up[i], h_sig_down[i], h_bkg_down[i], bins=bins, name=f'Systematic Variation {i+1}', save_name=f'Systematic_Variation_{i+1}')
```

```python
# read in data
with uproot.open("/your/path/to/data.root") as f:
    m = f['events']['m'].array()

# do fit
data_hist, _ = np.histogram(m, bins=bins)
minuit_obj = do_fit(data_hist, h_sig_nom, h_bkg_nom, h_sig_up, h_sig_down, h_bkg_up, h_bkg_down)

# plot fit result
plot_fit_results(
    data_hist, h_sig_nom, h_bkg_nom, h_sig_up, h_sig_down, h_bkg_up, h_bkg_down,
    minuit_obj, bins=np.linspace(60, 160, 21), save_name='Fit_Results'
)

# plot parameter values with errors
plot_fit_values(minuit_obj, save_name='Parameter_Values')
```
